```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Ad_Spend': [12000, 15000, 13000, 17000, 16000, 18000],  # Independent Variable
    'Revenue': [22000, 25000, 23000, 27000, 26000, 28000]     # Dependent Variable
}

# Create DataFrame
df = pd.DataFrame(data)

# Prepare features and target
X = df[['Ad_Spend']]
y = df['Revenue']

# Build Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Forecast future revenue
future_ad_spend = pd.DataFrame({'Ad_Spend': [19000, 20000]})
predicted_revenue = model.predict(future_ad_spend)

# Display predictions
for i, spend in enumerate(future_ad_spend['Ad_Spend']):
    print(f"Predicted revenue for Ad Spend ₹{spend}: ₹{predicted_revenue[i]:.2f}")

# Optional: Plot results
plt.scatter(X, y, color='blue', label='Actual Revenue')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Ad Spend')
plt.ylabel('Revenue')
plt.title('FMCG Revenue Forecasting')
plt.legend()
plt.grid(True)
plt.show()
```
